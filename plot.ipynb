```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
aris=np.loadtxt(fname='data/inflammation-01.csv',delimiter=',')
```

```python
image = plt.imshow(np.loadtxt(fname='data/inflammation-01.csv', delimiter=','))
```

```python
ave_infla=aris.mean(axis=0)
_=plt.plot(ave_infla)
plt.show()
```

```python
std_inf=aris.std(axis=0)
_=plt.plot(std_inf)
plt.show()
```

# subplot

```python
fig, ax = plt.subplots(3, 1, sharex='col', sharey='row')
```

```python
fig, ax = plt.subplots(3, 1, sharex='col', sharey='row',figsize=(10.0, 3.0))
ax[0].set_ylabel('average')
ax[0].plot(np.mean(aris, axis=0))

ax[1].set_ylabel('max')
ax[1].plot(np.max(aris, axis=0))

ax[2].set_ylabel('min')
ax[2].plot(np.min(aris, axis=0))
plt.show()
```
